```python
import requests
import bs4
import pandas as pd
```

```python
url = 'https://www.spotrac.com/redirect/player/102/' + 'injuries'
url
```

```python
res = requests.get(url)
soup=bs4.BeautifulSoup(res.text, 'html.parser')
```

```python
body_injury = soup.find('h2',text='Injuries').findNext('tbody')
body_injury
```

```python
tr = body_injury.find_all('tr')
```

```python
for t in tr:
    td = t.find_all('td')
    dates = td[0].text.split(' - ')
    print(dates[0], dates[1], td[1].text, td[2].text, td[3].text, td[4].text.replace('$','').replace(',',''))
```
