```python
import pandas as pd
import os
```

```python
cwd = os.getcwd().replace('\\', '/')
cwd
```

```python
columns = ['Rk','Name','Age','Tm',"G",'PA','Rbat','Rbaser','Rdp',\
           'Rfield','Rpos','RAA','WAA','Rrep','RAR','WAR','waaWL%',\
           '162WL%','oWAR','dWAR','oRAR','Salary','Acquired','Pos Summary',\
           'League','Year']
```

```python
columns = ['Rk','Name','Age','Tm',"G",'PA','Rbat','Rbaser','Rdp',\
           'Rfield','Rpos','RAA','WAA','Rrep','RAR','WAR','waaWL%',\
           '162WL%','oWAR','dWAR','oRAR','Salary','Acquired','Pos Summary',\
           'League','Year']

df = pd.DataFrame(columns=columns)


for year in range(2000,2020):
        for league in ('AL','NL'):
            df_add = pd.read_csv(cwd + '/' + league + "/" + league + '_' + str(year) + '_value_batting.csv')
            df_add['League'] = league
            df_add['Year'] = year
            df = pd.concat([df, df_add ], axis=0)
            

new = df['Name'].str.split('\\', n =1 , expand = True)


new[0] = new[0].str.replace('*','') 
df["name"]= new[0].str.replace('#','') 
  
df["name_key"] = new[1]

df["Salary"] = df["Salary"].str.replace('$','')

columns = ['name','name_key'] + columns[3:4] + ['League', 'Year'] + columns[4:-3]


df = df[columns]
df.head()

COLUMNS = []

for col in columns:
    COLUMNS.append(col.upper().replace(' ','_'))

df.columns = COLUMNS

destination = cwd.replace('Player Batting Value','')
df.to_csv(destination + 'player_value_batting.csv', index=False)
```
