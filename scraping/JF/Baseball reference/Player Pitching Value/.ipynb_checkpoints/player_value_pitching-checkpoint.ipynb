```python
import pandas as pd
import os
```

```python
cwd = os.getcwd().replace('\\', '/')
cwd
```

```python
columns = ['Rk','Name','Age','Tm','IP','G','GS','R','RA9','RA9opp','RA9def','RA9role',PPFp,RA9avg,RAA,WAA,gmLI,WAAadj,WAR,RAR,waaWL%,162WL%,Salary,Acquired]
```

```python
columns = ['Rk','Name','Age','Tm',"G",'PA','Rbat','Rbaser','Rdp',\
           'Rfield','Rpos','RAA','WAA','Rrep','RAR','WAR','waaWL%',\
           '162WL%','oWAR','dWAR','oRAR','Salary','Acquired','Pos Summary',\
           'League','Year']

df = pd.DataFrame(columns=columns)


for year in range(2000,2020):
        for league in ('AL','NL'):
            df_add = pd.read_csv(cwd + '/' + league + "/" + league + '_' + str(year) + '_value_batting.csv')
            df_add['League'] = league
            df_add['Year'] = year
            df = pd.concat([df, df_add ], axis=0)
            
columns = columns[:4] + ['League', 'Year'] + columns[4:-2]
df = df[columns[1:]]

destination = cwd.replace('Player Pitching Value','')
df.to_csv(destination + 'player_value_pitching.csv', index=False)
```
