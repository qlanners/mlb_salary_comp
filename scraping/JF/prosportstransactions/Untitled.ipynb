```python
import requests
import bs4
import pandas as pd
```

```python
from IPython.display import display_html
```

```python
url = "http://www.prosportstransactions.com/baseball/Search/SearchResults.php?Player=&Team=&BeginDate=2000-01-01&EndDate=2019-12-31&DLChkBx=yes&submit=Search"
```

```python
res = requests.get(url)
soup=bs4.BeautifulSoup(res.text, 'html.parser')
body = soup.find('table', {'class':'datatable center'})
```

```python
display_html(str(body), raw=True)
```

```python
def remove_dot(col):
    if not pd.isnull(col):
        return col.replace('• ','')
    else:
        return col
```

```python
dfs = pd.read_html(str(body), header=0)
df = dfs[0]
df['Acquired'] = df['Acquired'].apply(remove_dot)
df['Relinquished'] = df['Relinquished'].apply(remove_dot)
df
```
