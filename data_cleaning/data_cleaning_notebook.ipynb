```python
import pandas as pd
```

Checking Salaries first. 

```python
salaries = pd.read_csv('salaries.csv')
```

```python
salaries.head(10)
```

```python
salaries.drop(salaries[salaries['total_salary']=='-'].index, inplace=True)
salaries['total_salary'] = salaries['total_salary'].str.replace('$', '')
salaries['total_salary'] = salaries['total_salary'].str.replace(',', '')
```

```python
salaries.head(10)
```

```python
salaries.columns
```

```python
salaries = salaries.drop(['age', 'status', 'base_salary', 'signing_bonus', 'incentives',
        'adjusted_salary', 'payroll_perc', 'active', 'lux_tax',
        'team',], axis=1)
```

```python
salaries.head(10)
```

```python
inflation = pd.read_csv('Inflation.csv')
```

```python
inflation.head()
salaries.head()
```

```python
salaries = pd.merge(salaries,inflation, left_on = ['year'], right_on= ['year'] )
```

```python
salaries.head()
```

```python
salaries['total_salary'] = salaries['total_salary'].astype(int)
salaries['cpi'] = salaries['cpi'].astype(float)
```

```python
salaries['real_total_salary'] = salaries['total_salary'] * salaries['cpi']
```
