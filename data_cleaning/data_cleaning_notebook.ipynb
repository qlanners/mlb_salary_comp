So these two columns are not identical. How do they differ?

```python
counter = 0
indices = []
for t in zip(lst_adj_sal,lst_tot_sal):
    if t[0] != t[1]:
        indices.append((t[0],t[1],counter))
    counter += 1
indices
```

total_salaries is either equal or larger. So I will cut adjusted_salary.


Also, remove the lines where total_salary is null ('-'). Remove '$' as well.

```python
salaries.drop(salaries[salaries['total_salary']=='-'].index, inplace=True)
salaries['total_salary'] = salaries['total_salary'].str.replace('$', '')
salaries['total_salary'] = salaries['total_salary'].str.replace(',', '')
```
